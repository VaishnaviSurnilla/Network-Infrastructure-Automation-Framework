```python
import json
```

```python
# Sample network configuration
network_config = {
    "devices": [
        {"name": "Router1", "ip": "192.168.1.1", "vlan": 10},
        {"name": "Router2", "ip": "192.168.1.2", "vlan": 20}
    ]
}
```

```python
# Function to generate config
def generate_config(device):
    config = f"""
    hostname {device['name']}
    interface Gig0/0
    ip address {device['ip']} 255.255.255.0
    vlan {device['vlan']}
    no shutdown
    """
    return config
```

```python
# Deploy configs
for device in network_config['devices']:
    config = generate_config(device)
    print(f"Deploying config for {device['name']}")
    print(config)
```

```python
# Save configs
with open('configs.txt', 'w') as f:
    for device in network_config['devices']:
        f.write(generate_config(device))
```
